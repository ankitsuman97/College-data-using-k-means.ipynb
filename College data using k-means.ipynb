```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 2, random_state=90)
```

```python
#fit the model to all the data expect for the private label.
```

```python
km.fit(df.drop('Private', axis=1))
```

```python
##what are the cluster center vectors
```

```python
km.cluster_centers_
```

## Evalution

### There is no perfect way to evalution to evalute clustreing if you dont have the labels.

```python
def converter(prvt):
    if prvt== 'Yes':
        return 1
    else:
        return 0
    
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df
```

```python
#create a confusion matrics and classification_report to see hoe well the the kmeans clustering worked worked without
#being given any labels.
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'],km.labels_))
print(classification_report(df['Cluster'],km.labels_))
```
